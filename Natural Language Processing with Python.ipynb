#### Searching Text
Examine the context of a text in various ways other than simply reading it. One method is to look at the concordance of a word - this gives every occurance of the word, along with the context in which it appears.

Concordance simply allows us to view a given word in the context in which it appears in a body of text.

```python
text1.concordance("monstrous")
```

```python
text5.concordance("lol")
```

It is also possible to find out which other words appear in a similar context to a given word using the *similar* method

```python
# here we look for words that appread in the same context as the word monstrous
text1.similar("monstrous")
```

```python
# look at how another text uses it
text2.similar("monstrous")
```

Using the *similar* method we can see that the two authors use the word "monstrous" very differently. There is the *common_contexts* method that allows examination of contexts that are shared by two or more words. Note that this method requires a list of words as a parameter

```python
text2
```

```python
text2.common_contexts(["monstrous", "very"])
```

So far:
- detect if a word appears in a given text
- display words that appear in the same context

Also possible to detemermine the *location* of a word - i.e. how many words in from the beginning it occurs. This positional information can be displayed on a **dispersion plot**

```python
# import the dispersion plot function
from nltk.draw.dispersion import dispersion_plot
# create a list of words we want to analyse
words = ["citizens", "democracy", "freedome", "duties", "America"]

# call the dispersion plot function on a text, with a list of words
dispersion_plot(text4, words)
```

On the dispersion plot, each row (for each word) represents the entire body of text. Each vertical strip represents an occurance of the word, relative to the beginning of the text.

##### Tokens
A token is the technical name for a sequence of characters that we want to treat as a group. When we count the number of tokens in a text, we count the number of occurance of each unique sequence. For example:
*"to be or not to be"* has the following tokens, with associated occurances:
- to  : 2
- be  : 2
- or  : 1
- not : 1

The **vocabulary** of a text is just the *set* of **tokens** that make up the text. In a set, all duplicates are collapsed together.

```python
# In Python we can use the set() method to create a set of the text body, 
# this collpases duplicates and ensures that each element in the list is unique
set(text3)
```

```python
# sorted alphabetically
# punctuation symbols first
# all capitalized words precede lowercase words
sorted(set(text3))
```

```python
# combining this with the len() function gives us the number of tokens making of the text
len(set(text3))
```

The text therefore is made up of 44764 tokens, but only 2789 distict words or *word types*.
A **word type ** is the form or spelling of a word independent of its specific occurances in a text - i.e. a word considered to be a unique item of vocabulary.

```python
# on average, how many times is each word used
len(text3)/float(len(set(text3)))
```

So we can now consider the most common words and also the rarest words occuring in a text. What if this is still not even to make any inferences about the text?

Let's try looking at the ***long*** words of the text - remember, the goal is to find words to **characterize** the text.

```python
# lets look for words that are greater than 15 characters long
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

It is very common for these unsually long words to be **hapaxes**. Therefore they can't reliably typify a given text. So we've seen that short words *English Plumbin* occur often, and make up a large bulk of the body of text, but don't *say* anything about the text. A better approach then may be to look for ***frequently occuring long words***

```python
# get the frequency distribution
fdist5 = FreqDist(text5)
# create a list of frequently occuring long words
sorted([word for word in set(text5) if len(word) > 7 and fdist5[word] > 7])
```

With the above, we have managed to automatically identify the ***frequently occuring content-bearing*** words of the text

#### Collocations and Bigrams
A **Collocation** is a sequence of words that occur together unusually often. e.g. *red wine* A characteristic of collocations is that they are resistant to substitution with words that have similar senses. e.g. *maroon wine* - sounds odd

**Bigrams** are word pairs. Every possible pair of consecutive words is considered:
e.g.
*The quick brown fox* has bigrams:
- the quick
- quick brown
- brown fox

Collocations then are frequntly occuring bigrams, except we want to look for cases that involve rare words. In particular, we want to look at bigrams that occur more often than we would expect based on the frequency of individual words.

The *collocations* method does exactly this.

```python
# using the collocations function on text 4
text4.collocations()
```

```python
# using the collocations function on text 8
text8.collocations()
```

```python
text8
```

In the above we see how the collocations alone can give us some insight into the text.

#### Conditional Frequency Distributions
Frequency distributions compute the number of occurances of each item in a given list.
When a text of a corpus is divided into multiple categories, we can maintain separate frequency distributions for each category. This allows the studying of systematic differences between categories.

This can be achieved using NLTK's ConditionalFreqDis data type. A conditional frequency distribution is a collection of frequency distributions - one for each differen 'condition'.

Frequency distributions count the observable events - events are for example the appearance of a word. A conditional frequency distribution needs to pair each even with a condition. Therefore, we need to process a sequence of pairs.

e.g.

text = ['The', 'quick', 'brown', 'fox']

pairs = [('news', 'The'),  . . . ]

```python
from nltk.corpus import brown
# form the sequence of pairs as discusses
# pair in the forms of (condition, word/event)
genre_word = [(genre, word) for genre in ['news', 'romance'] for word in brown.words(categories=genre)]
```

```python
len(genre_word)
```

```python
len(brown.words(categories='news'))+len(brown.words(categories='romance'))
```

```python
# create the conditional frequency distribition
cfd = nltk.ConditionalFreqDist(genre_word)
```

```python
# check the conditions
cfd.conditions()
```

```python
# check the individual distributions
cfd['news']
```

```python
cfd['romance']
```

```python
cfd
```

```python
# using the individual distributions
cfd['news'].items()
```
